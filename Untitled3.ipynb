```python
import numpy as np
import random
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import scipy.io
import matplotlib.pyplot as plt
```

```python
seed = 20240111
np.random.seed(seed)
tf.random.set_seed(seed)
random.seed(seed)
```

```python
train_path = '/content/SVHN/train_32x32.mat'
test_path  = '/content/SVHN/test_32x32.mat'

train_data = scipy.io.loadmat(train_path)
test_data  = scipy.io.loadmat(test_path)

X_train_full = np.transpose(train_data['X'], (3,0,1,2)) / 255.0
y_train_full = train_data['y'].flatten()
y_train_full[y_train_full==10] = 0

X_test = np.transpose(test_data['X'], (3,0,1,2)) / 255.0
y_test = test_data['y'].flatten()
y_test[y_test==10] = 0

subset_size = 20000
X_train = X_train_full[:subset_size]
y_train = y_train_full[:subset_size]
```

```python
datagen = ImageDataGenerator(
    rotation_range=15,
    zoom_range=0.1,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True
)
datagen.fit(X_train)
```

```python
model = models.Sequential([
    layers.Conv2D(32,(3,3),activation='relu',input_shape=(32,32,3)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(128,activation='relu'),
    layers.Dense(10,activation='softmax')
])
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(datagen.flow(X_train,y_train,batch_size=64),
                    epochs=5,
                    validation_data=(X_test,y_test))
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'],label='Train')
plt.plot(history.history['val_accuracy'],label='Val')
plt.title("Accuracy"); plt.legend()
plt.subplot(1,2,2)
plt.plot(history.history['loss'],label='Train')
plt.plot(history.history['val_loss'],label='Val')
plt.title("Loss"); plt.legend()
plt.show()
```

```python
test_loss, test_acc = model.evaluate(X_test,y_test)
print("Test Accuracy:", test_acc)
```
